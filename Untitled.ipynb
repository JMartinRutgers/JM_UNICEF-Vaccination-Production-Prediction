```python
import pandas as pd
import numpy as np
```

```python
BCG = pd.read_csv('CSVs/vaccinations/BCG.csv', index_col=0)
DTP1 = pd.read_csv('CSVs/vaccinations/DTP1.csv', index_col=0)
DTP3 = pd.read_csv('CSVs/vaccinations/DTP3.csv', index_col=0)
HEPB3 = pd.read_csv('CSVs/vaccinations/HEPB3.csv', index_col=0)
HEPBB = pd.read_csv('CSVs/vaccinations/HEPBB.csv', index_col=0)
HIB3 = pd.read_csv('CSVs/vaccinations/HIB3.csv', index_col=0)
IPV1 = pd.read_csv('CSVs/vaccinations/IPV1.csv', index_col=0)
MCV1 = pd.read_csv('CSVs/vaccinations/MCV1.csv', index_col=0)
MCV2 = pd.read_csv('CSVs/vaccinations/MCV2.csv', index_col=0)
PCV3 = pd.read_csv('CSVs/vaccinations/PCV3.csv', index_col=0)
RCV1 = pd.read_csv('CSVs/vaccinations/RCV1.csv', index_col=0)
ROTAC = pd.read_csv('CSVs/vaccinations/ROTAC.csv', index_col=0)
YFV = pd.read_csv('CSVs/vaccinations/YFV.csv', index_col=0)
```

```python
allVac = pd.concat([BCG, DTP1, DTP3, HEPB3, HEPBB, HIB3, IPV1, MCV1, MCV2, PCV3, RCV1, ROTAC, YFV])
allVacs=allVac.replace(np.nan, '', regex=True)
allVacs
```
